```python
county_results=pd.read_csv("countypres_2000-2020.csv")
county_results
```

```python
def get_winner(df):
   df["Winner"]=df.sort_values(["candidatevotes"],ascending=False).head(1)["candidate"].squeeze()
   return df
```

```python
county_results["County"]=county_results["county_name"].str.strip()+", "+county_results["state_po"].str.strip()
df=county_results.groupby(["year","County"]).apply(get_winner)
```

# Final Dataset

```python
county_sentiment_biden_20=pd.read_csv("county_sentiment_biden_20.csv")
county_sentiment_biden_20.set_index("Unnamed: 0",inplace=True)
county_sentiment_trump_20=pd.read_csv("county_sentiment_trump_20.csv")
county_sentiment_trump_20.set_index("Unnamed: 0",inplace=True)
county_sentiment_trump_16=pd.read_csv("county_sentiment_trump_16.csv")
county_sentiment_trump_16.set_index("Unnamed: 0",inplace=True)
county_sentiment_hillary_16=pd.read_csv("county_sentiment_hillary_16.csv")
county_sentiment_hillary_16.set_index("Unnamed: 0",inplace=True)
county_sentiment_obama_12=pd.read_csv("county_sentiment_obama_12.csv")
county_sentiment_obama_12.set_index("Unnamed: 0",inplace=True)
county_sentiment_romney_12=pd.read_csv("county_sentiment_romney_12.csv")
county_sentiment_romney_12.set_index("Unnamed: 0",inplace=True)
```

```python
def get_final_dataset(county_sentiment_1,county_sentiment_2,county_results,GDP_data,year): #Keep first dataframe democrat and second dataframe republican
    df=pd.DataFrame()
    df["County"]=county_results["County"].str.lower().unique()
    df["Year"]=year
    if year==2020:
        df["GDP_YoY_%_Change"]=GDP_data["2020_2019_Change"]
    elif year==2016:
        df["GDP_YoY_%_Change"]=GDP_data["2016_2015_Change"]
    elif year==2012:
        df["GDP_YoY_%_Change"]=GDP_data["2013_2012_Change"]  # This is incorrect. Correct it.
    df=df.merge(county_sentiment_1, on="County").merge(county_sentiment_2, on="County").merge(county_results,on="County")
    df=df.get(["Year","County","Compound_Sentiment_x","Compound_Sentiment_y","Present_Term","GDP_YoY_%_Change","Winner"])
    df.rename(columns = {'Compound_Sentiment_x':'Sentiment_Republican', 'Compound_Sentiment_y':'Sentiment_Democrat'}, inplace = True)
    df["Winner"]=np.where((df["Winner"]=="DONALD TRUMP")|(df["Winner"]=="DONALD J TRUMP")|(df["Winner"]=="MITT ROMNEY"),1,0) 
    return df
```

```python
df=pd.DataFrame()
county_sentiment_trump_20["County"]=county_sentiment_trump_20["County"].str.lower()
county_sentiment_biden_20["County"]=county_sentiment_biden_20["County"].str.lower()
county_sentiment_hillary_16["County"]=county_sentiment_hillary_16["County"].str.lower()
county_sentiment_trump_16["County"]=county_sentiment_trump_16["County"].str.lower()
county_sentiment_obama_12["County"]=county_sentiment_obama_12["County"].str.lower()
county_sentiment_romney_12["County"]=county_sentiment_romney_12["County"].str.lower()
GDP_data["County"]=GDP_data["County"].str.lower()
county_results_2020["County"]=county_results_2020["County"].str.lower()
county_results_2016["County"]=county_results_2016["County"].str.lower()
county_results_2012["County"]=county_results_2012["County"].str.lower()

final_dataset_2012=get_final_dataset(county_sentiment_obama_12,county_sentiment_romney_12,county_results_2012,GDP_data,2012)
final_dataset_2016=get_final_dataset(county_sentiment_hillary_16,county_sentiment_trump_16,county_results_2016,GDP_data,2016)
final_dataset_2020=get_final_dataset(county_sentiment_biden_20,county_sentiment_trump_20,county_results_2020,GDP_data,2020)

final_dataset=pd.concat([final_dataset_2012, final_dataset_2016], ignore_index=True, sort=False)

final_dataset["Sentiment_Difference"]=(final_dataset["Sentiment_Republican"]-final_dataset["Sentiment_Democrat"])
final_dataset_2020["Sentiment_Difference"]=(final_dataset_2020["Sentiment_Republican"]-final_dataset_2020["Sentiment_Democrat"])
final_dataset
```

## Based on Sentiment Difference

```python
# There is only one dependent variable. Simply scale the variable between 0 and 1 to get prediction
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
prediction = scaler.fit_transform(final_dataset_2020['Sentiment_Difference'].values.reshape(-1,1))
prediction
```

```python
import matplotlib.pyplot as plt
plt.plot(prediction[:,0])
```

```python
plt.scatter(final_dataset_2020["Sentiment_Difference"],final_dataset_2020["Winner"])
plt.show()
```

```python
def get_classification(predictions,threshold):
    classes = np.zeros_like(prediction[:,0])
    for i in range(len(classes)):
        if predictions[i] > threshold:
            classes[i] = 1
    return classes
```

```python
from sklearn.metrics import confusion_matrix
def c_m_analysis(true,pred,threshold):
    tn, fp, fn, tp = confusion_matrix(true,get_classification(pred,threshold)).ravel()
    precision = tp/(tp+fp)
    recall = tp/(tp+fn)
    fpr = fp/(fp+tn)
    f_score = 2*precision*recall/(precision+recall)
    accuracy = (tp+tn)/(tp+tn+fp+fn)
    print("Precision:\t\t\t%1.2f identified as mines are mines"%(precision))
    print("Recall/TPR:\t\t\t%1.2f proportion of actual mines identified"%(recall))
    print("False Positive Rate:\t\t%1.2f proportion of rocks identified as mines"%fpr)
    print("f-score:\t\t\t%1.2f tradeoff between precision and recall"%(f_score))
    print("Accuracy:\t\t\t%1.2f how well the model has classified"%(accuracy))
    return accuracy

accuracy=c_m_analysis(final_dataset_2020["Winner"],prediction[:,0],0.5)
```

```python
threshold=[]
acc=[]
for thresh in np.arange(0,1,0.1):
    acc.append(c_m_analysis(final_dataset_2020["Winner"],prediction[:,0],thresh))
    threshold.append(thresh)
plt.plot(threshold,acc)
plt.show()
```
